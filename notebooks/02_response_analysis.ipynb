# Response Analysis

Analyse generated model responses and judge scores.

```python
import sys
sys.path.insert(0, '..')

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

responses = pd.read_csv('../outputs/responses/generated_responses.csv')
print(f'Total responses: {len(responses)}')
responses.head()
```

```python
# Response count per model and pipeline
pivot = responses.groupby(['model_name', 'pipeline_name']).size().unstack(fill_value=0)
pivot.plot(kind='bar', figsize=(12, 5))
plt.title('Response Count per Model & Pipeline')
plt.tight_layout()
plt.show()
```

```python
# Factuality scores
fact = pd.read_csv('../outputs/judge_scores/factuality_scores.csv')
sns.boxplot(data=fact, x='pipeline_name', y='factuality_score', hue='model_type')
plt.title('Factuality Score by Pipeline and Model Type')
plt.tight_layout()
plt.show()
```
