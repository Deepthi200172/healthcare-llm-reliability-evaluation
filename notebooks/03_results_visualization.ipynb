# Results Visualization

Produce thesis-ready figures from the final reliability scores.

```python
import sys
sys.path.insert(0, '..')

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('../outputs/final_results/model_pipeline_reliability_scores.csv')
print(df.to_string(index=False))
```

```python
# Final reliability score heatmap
pivot = df.pivot_table(
    index='model_name', columns='pipeline_name', values='final_reliability_score'
)
plt.figure(figsize=(10, 6))
sns.heatmap(pivot, annot=True, fmt='.3f', cmap='RdYlGn', vmin=0, vmax=1)
plt.title('Final Reliability Score (model × pipeline)')
plt.tight_layout()
plt.savefig('../outputs/final_results/heatmap_reliability.png', dpi=150)
plt.show()
```

```python
# Metric breakdown bar chart
metrics = ['factuality', 'hallucination_safety', 'consistency', 'robustness']
melted = df.melt(
    id_vars=['model_name', 'pipeline_name', 'model_type'],
    value_vars=metrics,
    var_name='metric',
    value_name='score',
)
g = sns.FacetGrid(melted, col='pipeline_name', height=5)
g.map_dataframe(
    sns.barplot, x='model_name', y='score', hue='metric', palette='Set2'
)
g.add_legend()
plt.suptitle('Metric Breakdown by Pipeline', y=1.02)
plt.tight_layout()
plt.savefig('../outputs/final_results/metric_breakdown.png', dpi=150)
plt.show()
```

```python
# Open-source vs Proprietary comparison
type_avg = df.groupby(['model_type', 'pipeline_name'])['final_reliability_score'].mean().reset_index()
sns.barplot(data=type_avg, x='pipeline_name', y='final_reliability_score', hue='model_type')
plt.title('Average Reliability: Open-source vs Proprietary')
plt.ylim(0, 1)
plt.tight_layout()
plt.savefig('../outputs/final_results/open_vs_proprietary.png', dpi=150)
plt.show()
```
