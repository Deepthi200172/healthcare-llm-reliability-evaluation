# Data Exploration — PubMedQA

Explore the 100-sample PubMedQA dataset used in this thesis.

```python
import sys
sys.path.insert(0, '..')

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('../data/processed/pubmedqa_sample_100.csv')
print(f'Shape: {df.shape}')
df.head()
```

```python
# Label distribution
df['final_decision'].value_counts().plot(kind='bar', title='Label Distribution')
plt.tight_layout()
plt.show()
```

```python
# Question length distribution
df['question_len'] = df['question'].str.split().str.len()
df['context_len'] = df['context'].str.split().str.len()

fig, axes = plt.subplots(1, 2, figsize=(12, 4))
df['question_len'].hist(ax=axes[0], bins=20)
axes[0].set_title('Question Length (words)')
df['context_len'].hist(ax=axes[1], bins=20)
axes[1].set_title('Context Length (words)')
plt.tight_layout()
plt.show()
```
